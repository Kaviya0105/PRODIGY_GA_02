```python
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image

# Function to generate image with Stable Diffusion
def generate_image_sd(prompt):
    model_id = "runwayml/stable-diffusion-v1-5"
    pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
    pipe = pipe.to("cuda")

    image = pipe(prompt).images[0]
    return image

# Prompt for image generation
prompt_sd = "A beautiful landscape with mountains and a river"

# Generate and display images
image_sd = generate_image_sd(prompt_sd)
image_sd.show()
image_sd.save("generated_image_sd.png")
```
